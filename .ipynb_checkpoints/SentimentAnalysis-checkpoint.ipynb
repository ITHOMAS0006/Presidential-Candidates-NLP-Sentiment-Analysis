```python
# Load comments
comments_df = pd.read_csv('comments_200.csv')

# Text Preprocessing function
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    # Convert to lowercase
    #text = text.lower()
    
    # Safely convert all values to lowercase, handling NaN and non-string values
    text = str(text) if pd.notnull(text) else ''
    text = text.lower()

    
    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenization
    tokens = text.split()
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    # # Lemmatization
    # lemmatizer = WordNetLemmatizer()
    # tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # # Stemming
    # stemmer = PorterStemmer()
    # tokens = [stemmer.stem(word) for word in tokens]
    return ' '.join(tokens)

# Apply preprocessing
comments_df['cleaned_comment'] = comments_df['comment'].apply(preprocess_text)
comments_df
```

```python
def identify_topic(comment):
    
    if isinstance(comment, str):
        comment_lower = comment.lower()
        mentions_harris = 'kamala harris' in comment_lower or 'harris' in comment_lower or 'kamala' in comment_lower or 'kamla' in comment_lower or 'joe biden' in comment_lower or 'biden' in comment_lower
        mentions_trump = 'donald trump' in comment_lower or 'trump' in comment_lower or 'donald' in comment_lower

        if mentions_harris and mentions_trump:
            return 'Both'
        elif mentions_harris:
            return 'Kamala Harris'
        elif mentions_trump:
            return 'Donald Trump'
    
    # Return 'Neither' if comment is not a string or doesn't mention either person
    return 'Neither'

# Apply topic identification to each comment
comments_df['topic'] = comments_df['comment'].apply(identify_topic)
comments_df
```

```python
import pandas as pd
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from transformers import pipeline
import nltk

# Assuming comments_df contains the data
# Filter out comments for topics other than "Donald Trump," "Kamala Harris," and "Both"
filtered_comments_df = comments_df[comments_df['topic'].isin(['Donald Trump', 'Kamala Harris', 'Both'])]

# Download VADER lexicon
nltk.download("vader_lexicon")

# Initialize models
vader_analyzer = SentimentIntensityAnalyzer()
transformer_sentiment = pipeline("sentiment-analysis", truncation=True, max_length=512)

# Helper function for Transformer model output
def get_transformer_sentiment(sentiment_score, neutral_threshold=0.1):
    """
    Process sentiment_score to classify it as 'positive', 'negative', or 'neutral'.
    Handles cases where sentiment_score might be empty or not in expected format.
    """
    # Check if sentiment_score is a list and has at least two elements
    if not sentiment_score or not isinstance(sentiment_score, list) or len(sentiment_score) < 2:
        # Return 'neutral' or a default value if sentiment_score is empty or invalid
        return 'neutral'

    # Assuming sentiment_score is a list with two elements [positive_score, negative_score]
    positive_score = sentiment_score[0]
    negative_score = sentiment_score[1]
    
    if abs(positive_score - negative_score) < neutral_threshold:
        return 'neutral'
    elif positive_score > negative_score:
        return 'positive'
    else:
        return 'negative'

# Define a threshold for neutrality
neutral_threshold = 0.05

# Apply sentiment analysis with the threshold for neutrality
filtered_comments_df['textblob_sentiment'] = filtered_comments_df['comment'].apply(
    lambda x: 'positive' if TextBlob(x).sentiment.polarity > neutral_threshold else
              ('negative' if TextBlob(x).sentiment.polarity < -neutral_threshold else 'neutral')
)

# Apply sentiment labeling with the threshold
filtered_comments_df['vader_sentiment'] = filtered_comments_df['comment'].apply(
    lambda x: 'positive' if vader_analyzer.polarity_scores(x)['compound'] > neutral_threshold else
              ('negative' if vader_analyzer.polarity_scores(x)['compound'] < -neutral_threshold else 'neutral')
)

# Apply transformer sentiment analysis with the threshold
filtered_comments_df['transformer_sentiment'] = filtered_comments_df['comment'].apply(
    lambda x: get_transformer_sentiment(transformer_sentiment(x), neutral_threshold=0.1)
)

# Separate the data for Trump and Harris for analysis
trump_comments = filtered_comments_df[filtered_comments_df['topic'] == 'Donald Trump']
harris_comments = filtered_comments_df[filtered_comments_df['topic'] == 'Kamala Harris']

# Function to calculate the distribution of sentiments
def sentiment_distribution(df, model_col):
    return df[model_col].value_counts(normalize=True) * 100

# Calculate distributions for each model for Trump comments
trump_textblob_dist = sentiment_distribution(trump_comments, 'textblob_sentiment')
trump_vader_dist = sentiment_distribution(trump_comments, 'vader_sentiment')
trump_transformer_dist = sentiment_distribution(trump_comments, 'transformer_sentiment')

# Calculate distributions for each model for Harris comments
harris_textblob_dist = sentiment_distribution(harris_comments, 'textblob_sentiment')
harris_vader_dist = sentiment_distribution(harris_comments, 'vader_sentiment')
harris_transformer_dist = sentiment_distribution(harris_comments, 'transformer_sentiment')

# Display distributions
print("Sentiment Distribution for Trump Comments")
print("TextBlob:", trump_textblob_dist)
print("VADER:", trump_vader_dist)
print("Transformers:", trump_transformer_dist)

print("\nSentiment Distribution for Harris Comments")
print("TextBlob:", harris_textblob_dist)
print("VADER:", harris_vader_dist)
print("Transformers:", harris_transformer_dist)

# Analysis summary
print("\nAnalysis Summary:")
print("The sentiment distribution shows the percentage of positive, negative, and neutral responses detected by each model.")
print("Comparing these distributions helps understand which model is more likely to label comments as positive, negative, or neutral for both topics.")
```

```python
import matplotlib.pyplot as plt

# Prepare data for plotting
models = ['TextBlob', 'VADER', 'Transformers']
sentiments = ['positive', 'negative', 'neutral']

# Trump comments sentiment distribution data
trump_data = {
    'TextBlob': trump_textblob_dist.reindex(sentiments).fillna(0),
    'VADER': trump_vader_dist.reindex(sentiments).fillna(0),
    'Transformers': trump_transformer_dist.reindex(sentiments).fillna(0)
}

# Harris comments sentiment distribution data
harris_data = {
    'TextBlob': harris_textblob_dist.reindex(sentiments).fillna(0),
    'VADER': harris_vader_dist.reindex(sentiments).fillna(0),
    'Transformers': harris_transformer_dist.reindex(sentiments).fillna(0)
}

# Plot Trump Sentiment Distribution
plt.figure(figsize=(10, 6))
for i, model in enumerate(models):
    plt.bar([x + i * 0.25 for x in range(len(sentiments))], trump_data[model], width=0.25, label=model)

plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.title('Sentiment Distribution for Trump Comments')
plt.xticks([x + 0.25 for x in range(len(sentiments))], sentiments)
plt.legend()
plt.show()

# Plot Harris Sentiment Distribution
plt.figure(figsize=(10, 6))
for i, model in enumerate(models):
    plt.bar([x + i * 0.25 for x in range(len(sentiments))], harris_data[model], width=0.25, label=model)

plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.title('Sentiment Distribution for Harris Comments')
plt.xticks([x + 0.25 for x in range(len(sentiments))], sentiments)
plt.legend()
plt.show()
```
