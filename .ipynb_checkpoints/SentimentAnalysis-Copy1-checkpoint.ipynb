```python
def fetch_video_ids_from_playlist(playlist_id):
    """Fetch video IDs from a playlist."""
    video_ids = []
    request = youtube.playlistItems().list(
        part="snippet",
        playlistId=playlist_id,
        maxResults=50
    )
    response = request.execute()
    
    while response:
        for item in response['items']:
            video_ids.append(item['snippet']['resourceId']['videoId'])
        
        # Check if there's another page of results
        if 'nextPageToken' in response:
            request = youtube.playlistItems().list(
                part="snippet",
                playlistId=playlist_id,
                maxResults=50,
                pageToken=response['nextPageToken']
            )
            response = request.execute()
        else:
            break

    return video_ids

def fetch_comments_from_videos(video_ids, comments_per_video=200, max_total_comments=15000):
    """Fetch up to comments_per_video comments for each video, with a maximum of max_total_comments overall.
       Ignore videos with disabled comments."""
    comments = []
    total_comment_count = 0  # Track total comments across all videos

    for video_id in video_ids:
        video_comment_count = 0  # Track comments per video
        comment_page_token = None

        while video_comment_count < comments_per_video and total_comment_count < max_total_comments:
            try:
                request = youtube.commentThreads().list(
                    part="snippet",
                    videoId=video_id,
                    maxResults=100,  # Fetch up to 100 comments per request
                    pageToken=comment_page_token,
                    textFormat="plainText"
                )
                response = request.execute()

                # Process each comment in the response
                for item in response.get('items', []):
                    comment_text = item['snippet']['topLevelComment']['snippet']['textDisplay']
                    comments.append({'video_id': video_id, 'comment': comment_text})
                    video_comment_count += 1
                    total_comment_count += 1

                    # Stop if we've reached the desired number of comments for this video
                    if video_comment_count >= comments_per_video or total_comment_count >= max_total_comments:
                        break

                # Check for the next page of comments for the current video
                comment_page_token = response.get('nextPageToken')
                if not comment_page_token:
                    break  # No more comments for this video

            except Exception as e:
                print(f"Skipping video {video_id} due to error: {e}")
                break  # Move to the next video if comments are disabled or another error occurs

        # Stop if the overall target of comments across all videos is reached
        if total_comment_count >= max_total_comments:
            break

    # Save to DataFrame
    comments_df = pd.DataFrame(comments)
    comments_df.to_csv('comments_200.csv', index=False)
    return comments_df

# Example usage
playlist_id = 'PLPmQXcocGF_Of7lqrqPfCwAcZwIk-cfr6'  # Replace with your YouTube playlist ID
video_ids = fetch_video_ids_from_playlist(playlist_id)
comments_df = fetch_comments_from_videos(video_ids, comments_per_video=200, max_total_comments=15000)
```

```python
# Load comments
comments_df = pd.read_csv('comments_200.csv')

# Text Preprocessing function
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    # Convert to lowercase
    #text = text.lower()
    
    # Safely convert all values to lowercase, handling NaN and non-string values
    text = str(text) if pd.notnull(text) else ''
    text = text.lower()

    
    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenization
    tokens = text.split()
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    # # Lemmatization
    # lemmatizer = WordNetLemmatizer()
    # tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # # Stemming
    # stemmer = PorterStemmer()
    # tokens = [stemmer.stem(word) for word in tokens]
    return ' '.join(tokens)

# Apply preprocessing
comments_df['cleaned_comment'] = comments_df['comment'].apply(preprocess_text)
comments_df
```

```python
def identify_topic(comment):
    
    if isinstance(comment, str):
        comment_lower = comment.lower()
        mentions_harris = 'kamala harris' in comment_lower or 'harris' in comment_lower or 'kamala' in comment_lower or 'kamla' in comment_lower or 'joe biden' in comment_lower or 'biden' in comment_lower
        mentions_trump = 'donald trump' in comment_lower or 'trump' in comment_lower or 'donald' in comment_lower

        if mentions_harris and mentions_trump:
            return 'Both'
        elif mentions_harris:
            return 'Kamala Harris'
        elif mentions_trump:
            return 'Donald Trump'
    
    # Return 'Neither' if comment is not a string or doesn't mention either person
    return 'Neither'

# Apply topic identification to each comment
comments_df['topic'] = comments_df['comment'].apply(identify_topic)
comments_df
```

```python
# Re-import necessary libraries and re-load the data due to environment reset
import pandas as pd
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from transformers import pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load the dataset again
file_path = 'filtered_comments.csv'
data = pd.read_csv(file_path)

# Filter rows where trump_sentiment or harris_sentiment is available and set up labels
sentiment_map = {'positive': 1, 'neutral': 0, 'negative': -1}
data_trump = data[data['trump_sentiment'] != 'not available'].copy()
data_trump['label'] = data_trump['trump_sentiment'].map(sentiment_map)

data_harris = data[data['harris_sentiment'] != 'not available'].copy()
data_harris['label'] = data_harris['harris_sentiment'].map(sentiment_map)

# Set up VADER and transformers sentiment analysis tools
vader_analyzer = SentimentIntensityAnalyzer()
transformer_analyzer = pipeline("sentiment-analysis")

# Define functions for TextBlob, VADER, and Transformer predictions
def textblob_sentiment(comment):
    blob = TextBlob(comment)
    polarity = blob.sentiment.polarity
    return 1 if polarity > 0 else (-1 if polarity < 0 else 0)

def vader_sentiment(comment):
    scores = vader_analyzer.polarity_scores(comment)
    compound = scores['compound']
    return 1 if compound > 0.05 else (-1 if compound < -0.05 else 0)

def transformer_sentiment(comment):
    result = transformer_analyzer(comment)[0]
    return 1 if result['label'] == 'POSITIVE' else -1 if result['label'] == 'NEGATIVE' else 0

# Apply sentiment analysis on both Trump and Harris datasets
data_trump['textblob_pred'] = data_trump['cleaned_comment'].apply(textblob_sentiment)
data_trump['vader_pred'] = data_trump['cleaned_comment'].apply(vader_sentiment)
data_trump['transformer_pred'] = data_trump['cleaned_comment'].apply(transformer_sentiment)

data_harris['textblob_pred'] = data_harris['cleaned_comment'].apply(textblob_sentiment)
data_harris['vader_pred'] = data_harris['cleaned_comment'].apply(vader_sentiment)
data_harris['transformer_pred'] = data_harris['cleaned_comment'].apply(transformer_sentiment)

# Define function to calculate evaluation metrics
def evaluate_sentiment_models(data, model_pred_cols):
    results = {}
    for model, pred_col in model_pred_cols.items():
        # Drop rows with NaN values in either 'label' or the prediction column
        filtered_data = data.dropna(subset=['label', pred_col])
        
        # Calculate metrics
        accuracy = accuracy_score(filtered_data['label'], filtered_data[pred_col])
        precision = precision_score(filtered_data['label'], filtered_data[pred_col], average='weighted', zero_division=0)
        recall = recall_score(filtered_data['label'], filtered_data[pred_col], average='weighted', zero_division=0)
        f1 = f1_score(filtered_data['label'], filtered_data[pred_col], average='weighted', zero_division=0)
        
        # Store results
        results[model] = [accuracy, precision, recall, f1]
    return pd.DataFrame(results, index=['Accuracy', 'Precision', 'Recall', 'F1 Score'])

# Evaluation for Trump comments
trump_results_df = evaluate_sentiment_models(data_trump, {
    'TextBlob': 'textblob_pred',
    'VADER': 'vader_pred',
    'Transformer': 'transformer_pred'
})

# Evaluation for Harris comments
harris_results_df = evaluate_sentiment_models(data_harris, {
    'TextBlob': 'textblob_pred',
    'VADER': 'vader_pred',
    'Transformer': 'transformer_pred'
})

print("Trump Sentiment Analysis Results")
print(trump_results_df)

print("Harris Sentiment Analysis Results")
print(harris_results_df)
```
